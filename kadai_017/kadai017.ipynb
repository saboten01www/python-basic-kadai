```python
class Human:
  def __init__(self,name,age):
     self.name = name
     self.age = age
  def check_adult(self):
    if self.age >= 20:
      print(f"{self.name} は成人です。")
    else:
      print(f"{self.name} は未成年です")

# Create multiple Human instances
taro = Human("侍太郎", 21)
hanako = Human("花子", 18)
jyuji = Human("十次", 30)

# Add instances to a list
humans = [taro, hanako, jyuji]

# Call check_adult for each instance
for human in humans:
  human.check_adult()
```
