```python
#ランダムな整数を利用するために、randomモジュールをインポート
import random

# 変数varに1～15までのランダムな整数を代入する
var = random.randint(1, 15)

# 変数numの値を出力
print(var)

# 変数varが、3の倍数と5の倍数の両方に該当する場合は「FizzBuzz」を出力
if var % 3 == 0 and var % 5 == 0 :
    print("FizzBuzz")
    #変数varが、3の倍数に該当する場合は「Fizz」を出力
elif var % 3 == 0:
    print("Fizz")
    #変数varが、5の倍数に該当する場合は「Buzz」を出力
elif var % 5 == 0:
    print("Buzz")
else:
    # それ以外の場合は、変数varの値を出力
    print(var)
```
